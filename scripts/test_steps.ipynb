```python
#fixing '5E54' becoming '5.00E+54' issue
def fix_PDB_ID(df):
    for i, j in enumerate(df['PDB_ID']):
        print (j)
        if len(str(j))>4:
            if '-' in j:
                X= ''.join(j.split('-')).upper()
                df.loc[i, 'PDB_ID']= X
            else:
                if '.' in j:
                    x= str(j)
                    x1= x.split('.')[0]
                    x2= x.split('+')[1]
                    X= x1+'E'+x2
                    print (X)
                    df.loc[i, 'PDB_ID']= X
                else:
                    x= str(j)
                    x1= x.split('+')[0]
                    x2= x.split('+')[1]
                    X= x1+ x2
                    print (X)
                    df.loc[i, 'PDB_ID']= X
                    
    return (df)
```

```python
def string_split(a):
    #function to extract chain ID (chain), residue name (resname), residue index (resnum), and icode
    #input (a) format: a.U1007 -->(chain ID).(residue ID)(residue index)
    b= list(a.split('.')[1])
    chain = list(a.split('.'))[0]
    resname = ''
    resnum = ''
    icode = '' #add an empty space here when this function is used to extract RSRZ values from RCSB dataset
    
    for i in range(len(b)):
        if b[i].isalpha():
            resname += b[i]
        else:
            break
        
    c= ''.join(b)
    d= c[len(resname):] #anything other than residue ID
    
    for i in range(len(d)):
        if d[i].isalpha():
            icode += d[i]
        else:
            resnum += d[i]
    
    if len(icode) ==1:
        pass
    else:
        icode = icode.replace(' ', '')
    #icode = icode.replace(' ', '')
    D={}
    D['chain'] = chain
    D['resname'] = resname
    D['resnum'] = resnum
    D['icode'] = icode
    
    return D
```
